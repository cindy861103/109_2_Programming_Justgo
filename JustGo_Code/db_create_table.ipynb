### Add PK & FK

```python
add_key = '''
    ALTER TABLE member ADD PRIMARY KEY (student_id);
    ALTER TABLE publish_m ADD PRIMARY KEY (pub_id);
    ALTER TABLE publish_d ADD PRIMARY KEY (pub_d_id);
    ALTER TABLE publish_d ADD FOREIGN KEY (pub_id) REFERENCES publish_m (pub_id);
    ALTER TABLE publish_d ADD FOREIGN KEY (student_id) REFERENCES member (student_id);
    ALTER TABLE order_m ADD PRIMARY KEY (order_id);
    ALTER TABLE order_m ADD FOREIGN KEY (pub_id) REFERENCES publish_m (pub_id);
    ALTER TABLE order_d ADD PRIMARY KEY (order_d_id);
    ALTER TABLE order_d ADD FOREIGN KEY (order_id) REFERENCES order_m (order_id);
    ALTER TABLE order_d ADD FOREIGN KEY (student_id) REFERENCES member (student_id);
    ALTER TABLE order_cancel ADD FOREIGN KEY (order_d_id) REFERENCES order_d (order_d_id);
    ALTER TABLE order_cancel ADD FOREIGN KEY (order_id) REFERENCES order_m (order_id);
    ALTER TABLE order_cancel ADD FOREIGN KEY (student_id) REFERENCES member (student_id);
'''
cursor.execute(add_key)
conn.commit()
print('Add Key Successfully!!')
```
